```python
import pandas as pd
import glob2
import os
```

```python
path = ".."
files = glob2.glob(f'{path}/real_data/*.h5')
files = [f[len(f"{path}/real_data/"):-3] for f in files]
print(files)
```

```python
category = "real_data"
# category = "balanced_data"
# category = "imbalanced_data"
# os.listdir(f"results/{category}")
```

```python
df = pd.DataFrame(columns = ["dataset", "method", "run", "ARI", "NMI", "pred"])
for f  in os.listdir(f"results/{category}"):
    dataset = f.split(".h5")[0]
    method = f.split(".")[-2].split("_")[1]
    run = f.split(".")[-2].split("_")[2]
    d =pd.read_csv(f"results/{category}/{f}")
    values =d.values[:, 1]

    ari = values[0]
    nmi = values[1]
    pred = values[2:]
    df.loc[df.shape[0]] = [dataset, method, run, ari, nmi, pred]
    

df
```

```python
df.groupby(["dataset", "method"])["ARI"].mean().unstack("dataset")
```

```python
df.to_pickle(f"{category}.pkl")
```
